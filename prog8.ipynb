```python
# Translation

import cv2
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('fru.png')
image_mat= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

ht,wd,c=image.shape

tx=int(input("Enter the value of tx: "))
ty=int(input("Enter the value of ty: "))

trans=np.float32([[1,0,tx],[0,1,ty]])

image_t=cv2.warpAffine(image_mat,trans,(wd+500,ht+500))

plt.imshow(image_t)
```

```python
# Rotation

import cv2
import math
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('fru.png')
image_mat= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

ht,wd,c=image.shape

theta=int(input("Enter the value of theta: "))
angle=(theta*3.14)/180
cos=math.cos(angle)
sin=math.sin(angle)

rota=np.float32([[cos,-sin,0],[sin,cos,0]])

image_r=cv2.warpAffine(image_mat,rota,(wd+100,ht+100))

plt.imshow(image_r)
```

```python
# Rotation at a pivort point

import cv2
import math
import numpy as np
import matplotlib.pyplot as plt
image = cv2.imread('fru.png')
image_mat= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

ht,wd,c=image.shape

theta=int(input("Enter the value of theta: "))
angle=(theta*3.14)/180
cos=math.cos(angle)
sin=math.sin(angle)

xf=int(input("Enter the value of fixed X: "))
yf=int(input("Enter the value of fixed Y: "))

rota=np.float32([[cos,-sin,xf*(1-cos)+yf*sin],[sin,cos,yf*(1-cos)+xf*sin]])

image_r=cv2.warpAffine(image_mat,rota,(wd+1000,ht+1000))

plt.imshow(image_r)
```

```python
# every image in line

import cv2
import numpy as np
import matplotlib.pyplot as plt
image=cv2.imread('fru.png')
image_mat=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

ht, wd, c =image.shape
center =(wd/2, ht/2)

#Translation
tx=int(input("Enter the value of tx: "))
ty=int(input("Enter the value of ty: "))
trans=np.float32([[1,0,tx],[0,1,ty]])
#Scaling
sx=int(input("Enter the value of sx: "))
sy=int(input("Enter the value of sy: "))
scale=np.float32([[sx,0,0],[0,2,0]])
#Rotation
theta=int(input("Enter the value of theta: "))
angle=(theta*3.14)/180
cos=math.cos(angle)
sin=math.sin(angle)
rota=np.float32([[cos,-sin,0],[sin,cos,0]])


image_t= cv2.warpAffine(image_mat, trans, (wd +50, ht + 50))
image_s = cv2.warpAffine(image_mat, scale, (wd, ht))
image_r= cv2.warpAffine(image_mat, rota, (wd, ht)) 

fig,axs=plt.subplots(1,4)
  
axs[0].imshow(image_mat)    
axs[1].imshow(image_t)   
axs[2].imshow(image_s)
axs[3].imshow(image_r)   
```
