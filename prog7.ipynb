```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread('fru.png')
image_mat= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

ht,wd,c=image.shape
midx=wd//2
midy=ht//2

tl=image_mat[:midy,:midx]
tr=image_mat[:midy,midx:]
bl=image_mat[midy:,:midx]
br=image_mat[midy:,midx:]

fig,axs=plt.subplots(2,2)

axs[0,0].imshow(tl)
axs[0,0].axis('off')
axs[0,0].set_title("Top Left")

axs[0,1].imshow(tr)
axs[0,1].axis('off')
axs[0,1].set_title("Top Right")

axs[1,0].imshow(bl)
axs[1,0].axis('off')
axs[1,0].set_title("Bottom Left")

axs[1,1].imshow(br)
axs[1,1].axis('off')
axs[1,1].set_title("Bottom Right")


plt.axis("off")
plt.show()
```

```python
# Using for LOOP

import cv2
import matplotlib.pyplot as plt
image = cv2.imread('fru.png')
image_mat= cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

ht,wd,c=image.shape
midx=wd//2
midy=ht//2

tl=image_mat[:midy,:midx]   # : means all rows from 0 to midy and all columns from 0 to midx
tr=image_mat[:midy,midx:]
bl=image_mat[midy:,:midx]
br=image_mat[midy:,midx:]

fig,axs=plt.subplots(2,2)

l_title=["Top Left","Top Right","Bottom Left","Bottom Right"]
l_var=[tl,tr,bl,br]
k=0

for i in range(2):
    for j in range(2):
        axs[i,j].imshow(l_var[k])
        axs[i,j].axis("off")
        axs[i,j].set_title(l_title[k])
        k=k+1
        
plt.show()
```
